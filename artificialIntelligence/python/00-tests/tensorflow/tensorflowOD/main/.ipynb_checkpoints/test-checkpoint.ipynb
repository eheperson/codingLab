```python
import cv2
import uuid
import time
import os
```

```python
CLASSES = ['sunglas', 'pen', 'smartphone', 'notebook', 'glass']
numInstance = 5
DATA_PATH = os.path.join('workspace', 'dataset')
LABELIMG_PATH = os.path.join('workspace', 'labelimg')

TRAIN_PATH = os.path.join('workspace', 'images', 'train')
TEST_PATH = os.path.join('workspace', 'images', 'test')
ARCHIVE_PATH = os.path.join('workspace', 'images', 'archive.tar.gz')
```

```python
if not os.path.exists(DATA_PATH):
    if os.name == 'posix':
        !mkdir -p {DATA_PATH}
    if os.name == 'nt':
         !mkdir {DATA_PATH}
```

```python
for label in CLASSES:
    path = os.path.join(DATA_PATH, label)
    if not os.path.exists(path):
        !mkdir {path}
```

```python
for label in CLASSES:
    cap = cv2.VideoCapture(0)
    print('Collecting images for {}'.format(label))
    time.sleep(5)
    for imgnum in range(numInstance):
        print('Collecting image {}'.format(imgnum+1))
        
        while(1):
            ret, frame = cap.read()
            cv2.imshow('frame', frame)
            if cv2.waitKey(1) & 0xFF == ord('w'):
                break
        
        imgname = os.path.join(DATA_PATH,label,label+'.'+'{}.jpg'.format(str(uuid.uuid1())))
        cv2.imwrite(imgname, frame)
        time.sleep(2)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
cap.release()
cv2.destroyAllWindows()
```
