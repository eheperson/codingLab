```python
import numpy as np
from random import randint
from sklearn.preprocessing import MinMaxScaler
```

```python
train_labels = []
train_samples = []
```

<h4>Example Data : </h4>
<ul>
    <li>An experimental drug was tested on individuals from ages 13 to 65</li>
    <li>The trial had 2100 participants. Half were under 65 years old, half were over 65 years old.</li>
    <li>95% of patients 65 or older experienced side effects.</li>
    <li>95% of patients under 65 experienced no side effects.</li>
</ul>


```python
for i in range (1000):
    random_younger = randint(13,64)
    train_samples.append(random_younger)
    train_labels.append(0)
    
    random_older = randint(65,100)
    train_samples.append(random_older)
    train_labels.append(1)
    
for i in range(50):
    random_younger = randint(13,64)
    train_samples.append(random_younger)
    train_labels.append(1)
    
    random_older = randint(65,100)
    train_samples.append(random_older)
    train_labels.append(0)
```

```python
# print raw data
for i in train_samples:
    print(i)
```

```python
train_labels = np.array(train_labels)
train_samples = np.array(train_samples)
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
scaled_train_samples = scaler.fit_transform((train_samples).reshape(-1,1))
```

```python
#print scaled data
for i in scaled_train_samples:
    print(i)
```

```python
import keras
from keras import backend as K
from keras.models import Sequential
from keras.layers import Activation
from keras.layers.core import Dense
from keras.optimizers import Adam
from keras.metrics import categorical_crossentropy
```

```python
# model = Sequential([laye1, layer2, layer3])
# or
# model.add(layer4)

model = Sequential([
    Dense(16,input_shape=(1,), activation="relu"),
    Dense(32, activation="relu"),
    Dense(2, activation="softmax")
])
```

```python
model.summary()
```

```python
model.compile(Adam(lr=.0001), loss="sparse_categorical_crossentropy", metrics=["accuracy"])
```

```python
model.fit(scaled_train_samples, train_labels, validation_split=0.1, batch_size=10, epochs=20, shuffle=True, verbose=2)
```

<h2> Predictions </h2>

```python
test_labels = []
test_samples = []
```

```python
for i in range(10):
    #the 5% of younger individuals who did experience side effects
    random_younger = randint(13,64)
    test_samples.append(random_younger)
    test_labels.append(1)
    
    #the 5% of older individuals who did not experience side effects
    random_older = randint(64,100)
    test_samples.append(random_older)
    test_labels.append(0)

for i in range(200):
    #the 95% of younger individuals who did experience side effects
    random_younger = randint(13,64)
    test_samples.append(random_younger)
    test_labels.append(1)
    
    #the 95% of older individuals who did not experience side effects
    random_older = randint(64,100)
    test_samples.append(random_older)
    test_labels.append(0)
    
```

```python
test_labels = np.array(train_labels)
test_samples = np.array(train_samples)
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
scaled_test_samples = scaler.fit_transform((test_samples).reshape(-1,1))
```

```python
predictions = model.predict(scaled_test_samples, batch_size=10, verbose=0)
```

```python
for i in predictions:
    print(i)
```

```python
rounded_predictions = model.predict_classes(scaled_test_samples, batch_size=10, verbose=0)
```

```python
for i in rounded_predictions:
    print(i)
```

Create Confusion Matrix

```python
%matplotlib inline
from sklearn.metrics import confusion_matrix
import itertools
import matplotlib.pyplot as plt
```

```python
cm = confusion_matrix(test_labels, rounded_predictions)
```

```python
def plotCm(cm,
          classes,
          normalize=False,
          title='confusion_matrix',
          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting 'normalize=True'.
    """
    
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes,rotation=45)
    plt.yticks(tick_marks, classes)
    
    if normalize:
        cm = cm.astype('float')/ cm.sum(axis=1)[:,np.newaxis]
        print("Normalized Confusion MAtrix")
    else:
        print("COnfusion Matrix, Without Normalization")
        
    print(cm)
    
    thresh = cm.max() / 2.
    
    for i,j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i,j],
                horizontalalignment = 'center',
                color='white' if cm[i, j] > thresh else "black")
        
    plt.tight_layout()
    plt.ylabel("True Label")
    plt.xlabel("Predicted Label")
    
```

```python
cm_plot_labels = ['no_side_effects', 'half_side_effects']

plotCm(cm, cm_plot_labels, title='Confusion Matrix')
```

Save And Load Model

The Save functions saves :
    1- The architecture of the model allowing to re-create the model
    2- the weights of the model
    3- the training configurations(loss, optimizer)
    4- the state of the optimizer, allowing to resume training exactly where you are left off

```python
#Saving Model
model.save('medical_trial_model.h5')

#Load Model
from keras.models import load_model
new_model = load_model('medical_trial_model.h5')
```
